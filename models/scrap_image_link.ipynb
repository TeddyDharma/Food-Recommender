```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.request
from main import *
import json
from io import BytesIO
from PIL import Image
```

```python


def get_data_from_json_file(json_data): 
    return json.loads(json_data)['recommendation']


def read_data(): 
    data = pd.read_excel("../data/recommendartion_data/finalIndonesia_data.xlsx")
    return data
def get_image_link(data : pd.DataFrame, attr : list, class_atr: str, iter = 1):
    arr = []
    for image_mame in data[:100]: 
        try: 
            url = requests.get(f'https://www.bing.com/images/search?q={image_mame}&form=HDRSC3')
            soup = BeautifulSoup(url.content, 'lxml')
            get_image_link = soup.find(name = attr[0], attrs={"class" : class_atr})
            get_href = get_image_link.find(attr[1], href=True)['href']
            final = "https://www.bing.com/" + get_href
            arr.append(final)
        except Exception as e: 
            print(f'error : {e} or image tag not detected')
    return arr

def get_second_links(data_link):
    arr_image_link = []

    for link in data_link:
        try: 
            url = requests.get(link)
            soup = BeautifulSoup(url.content, 'lxml') 
     
            get_image_link = soup.find_all(name = "div",id = "b_content")
          
            if get_image_link: 
  
                for attributes in str(get_image_link).split(":"):
                    for j in str(attributes).split(":"):
                        # print(j)
                        if "ImgDet" and ".mm.bing.net/th/id/OIP." in j:
                            arr_image_link.append("https:" + j.replace("fhs", "").replace('"' , "").split(",")[0])
                         
        except Exception as e: 
            print(f'error : {e} or image tag not detected')
    return arr_image_link


# second_get = get_image_link()
```

```python
def download_images(array_links, food_name): 
      for i in range(0, len(array_links)):
        response = requests.get(array_links[i])
        print(f'getting {food_name[i]} using  {array_links[i]}')
        try:
            if os.path.exists("./recommendation_images"):
                images_lod = Image.open(BytesIO(response.content))
                full_path = f'./recommendation_images/{food_name[i]}.png'
                images_lod.save(full_path)
            else: 
                os.mkdir("./recommendation_images")
        except Exception as e: 
            print('cant get the image')
        
        
```
