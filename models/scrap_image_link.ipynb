```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python
def read_data(): 
    data = pd.read_excel("../data/recommendartion_data/finalIndonesia_data.xlsx")
    return data
def get_image_link(data : pd.DataFrame, attr : list, class_atr: str, iter = 1):
    arr = []
    for image_mame in data.loc[:10, 'nama_makanan']: 
        try: 
            url = requests.get(f'https://www.bing.com/images/search?q={image_mame}&form=HDRSC3')
            soup = BeautifulSoup(url.content, 'lxml')
            get_image_link = soup.find(name = attr[0], attrs={"class" : class_atr})
            get_href = get_image_link.find(attr[1], href=True)['href']
            final = "https://www.bing.com/" + get_href
            # print(final)

            arr.append(final)
        
        except Exception as e: 
            print(f'error : {e} or image tag not detected')
    return arr

def get_second_links(data_link):
    for link in data_link:
        try: 
            url = requests.get(link)
            soup = BeautifulSoup(url.content, 'lxml') 
            # print(soup)
            # print("==========")
            get_image_link = soup.find_all(name = "div",id = "b_content")
            # get_src = get_image_link
            # print(get_image_link)
            if get_image_link: 
                for attributes in str(get_image_link).split(":"):
                    print(attributes) 

        except Exception as e: 
            print(f'error : {e} or image tag not detected')


# second_get = get_image_link()
```
